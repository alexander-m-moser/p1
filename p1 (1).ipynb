```python
#Q17
years = {}
for name in os.listdir("home-computers"):
    if name.endswith(".csv"):
        year = int(name[name.index("2"):name.index(".")])
        years[year] = pd.read_csv(os.path.join("home-computers", name), skiprows=1)
        years[year].set_index('Geographic Area Name', inplace = True, drop = True)
    else:
        continue
my_list = list(years.keys())
my_list.sort()
my_list
```

```python
#Q18
df = years[2018]
df.at["Wisconsin", "Estimate!!Total"]
```

```python
#Q19
df["Estimate!!Total!!No Computer"].sum()
```

```python
#Q20
target_val = (df["Margin of Error!!Total!!No Computer"] / df["Estimate!!Total!!No Computer"]).max()
target_val
```

```python
#Q21

illinois = df.at["Illinois", "Estimate!!Total!!No Computer"] / df.at["Illinois", "Estimate!!Total"]
indiana = df.at["Indiana", "Estimate!!Total!!No Computer"] / df.at["Indiana", "Estimate!!Total"]
iowa = df.at["Iowa", "Estimate!!Total!!No Computer"] / df.at["Iowa", "Estimate!!Total"]
michigan = df.at["Michigan", "Estimate!!Total!!No Computer"] / df.at["Michigan", "Estimate!!Total"]
minnesota = df.at["Minnesota", "Estimate!!Total!!No Computer"] / df.at["Minnesota", "Estimate!!Total"]
wisconsin = df.at["Wisconsin", "Estimate!!Total!!No Computer"] / df.at["Wisconsin", "Estimate!!Total"]

state_values = {"Illinois" : illinois * 100, "Indiana" : indiana * 100, "Iowa" : iowa* 100, 
                "Michigan" : michigan * 100, "Minnesota" : minnesota *100, "Wisconsin" : wisconsin * 100}

state_values
```

```python
#Q22
ser = pd.Series(state_values)
ser.plot.bar()
```

```python
#Q23

wi ={}
for year in my_list:
    wi[year] = (years[year].at["Wisconsin", "Estimate!!Total!!No Computer"]/1000)

answer = pd.Series(wi)
answer.plot.line().set_xlabel("Year")
answer.plot.line().set_ylabel("WI households without Computers (thousands)")
```
